```python
import datetime
from os import listdir
from os.path import join, exists
from shutil import copyfile
from pandas import ExcelFile

path_to_raw_directory = 'C:/Users/b_charmettant/data/parotide_ml_raw/'
path_to_directory = 'C:/Users/b_charmettant/data/parotide_ml/'
```

```python
def get_date_exam(path):
    with open(path, 'r') as f:
        data = f.readlines()
    return format_date(data[6].split(';')[1])
    

def igr_to_parotide(id_igr, date, path_to_overview='C:/Users/b_charmettant/data/parotide_ml_raw/'):
    overview = ExcelFile(join(path_to_overview, 'overview_complet.xlsx')).parse(0).values
    for line in overview[1:]:
        date_line = line[5].date()
        
        
        if line[1].replace(' ', '').lower() == id_igr.lower():
            if date_line == date:
                return line[0]

    raise NameError(f'{id_igr} at {date} couldn\'t be found')


def format_date(str_date):
    month_dict = {'Jan':'01', 'Feb':'02', 'Mar':'03', 'Apr':'04', 'May':'05', 'Jun':'06',
                  'Jul':'07','Aug':'08','Sep':'09','Oct':'10','Nov':'11','Dec':'12'}
    
    date = str_date.split(' ')
    
    return datetime.date(int(date[0]), int(month_dict[str(date[1])]), int(date[2]))

def image_type(name_image):
    suffix = name_image.split('_')[-1]
    
    if 't2' in suffix.lower():
        return 'T2'
    elif 't1' in suffix.lower():
        return 'T1'
    elif 'gado' in suffix.lower():
        return 'GADO'
    elif 'diff' in suffix.lower():
        return 'DIFF'
    elif 'dwi' in suffix.lower():
        return 'DIFF'
    else:
        raise NameError(f"{name_image} couldn\'t be classified")
```

```python
ls_files = listdir(join(path_to_raw_directory, 'exams'))

for file in ls_files:
    src = join(join(path_to_raw_directory, 'exams'), file)
    date = get_date_exam(src)
    file_1 = file.split('_', 1)
    parotide_id = igr_to_parotide(file_1[0], date)
    
    file_type = image_type(file)
    
    new_file = f"{parotide_id}_{file_type}.csv"

    tgt = join(join(path_to_directory, 'exams'), new_file)
    
    if exists(tgt):
        print(f"{new_file} already exists ({file})")
    else:
        print(new_file)
        copyfile(src, tgt)
    
print("Done !")
```
